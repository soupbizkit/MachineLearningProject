```python
#INTEGRANTES:
#Walter Santacruz - 1630645
#Andres Valencia - 
#Santiago Quintero -

import numpy as np # linear algebra
import matplotlib.pyplot as plt
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os

data = pd.read_csv('heart.csv') #leer datos
print(data)
```

# Gráficos variables numéricas

```python
keyListNumeric = ['age', 'trestbps','chol','thalach', 'oldpeak',]


histogramas = plt.figure(figsize=(30, 20))

index = 1
for key in keyListNumeric:
    axes = histogramas.add_subplot(3,5,index)
    data[[key]].plot.hist(ax=axes, rwidth=1, edgecolor='black')
    index+= 1
    
```

# Gráficos variables no numéricas

```python
keyListNonNumeric = ['sex', 'fbs', 'restecg', 'exang', 'thal', 'slope', 'ca', 'cp' ];

histogramas = plt.figure(figsize=(30, 10))

index = 1
for key in keyListNonNumeric:
    axes = histogramas.add_subplot(2,4,index)
    data[[key]].value_counts().plot.pie(startangle=90, ylabel=key, fontsize=15)
    index+= 1
```
