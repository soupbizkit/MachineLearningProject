```python
#INTEGRANTES:
#Walter Santacruz - 1630645
#Andres Valencia - 
#Santiago Quintero -

import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import pandas as pd 
import os

##lectura de nuestros datos como dataFrame de pandas
data = pd.read_csv('heart.csv') #leer datos
print(data)
```

# Gráficamos las variables numéricas

```python
keyListNumeric = ['age', 'trestbps','chol','thalach', 'oldpeak',]


histogramas = plt.figure(figsize=(30, 20))

index = 1
for key in keyListNumeric:
    axes = histogramas.add_subplot(3,5,index)
    data[[key]].plot.hist(ax=axes, rwidth=1, edgecolor='black')
    index+= 1
    
```

# Gráficamos variables no numéricas

```python
keyListNonNumeric = ['sex', 'fbs', 'restecg', 'exang', 'thal', 'slope', 'ca', 'cp' ];

histogramas = plt.figure(figsize=(30, 10))

index = 1
for key in keyListNonNumeric:
    axes = histogramas.add_subplot(2,4,index)
    data[key].value_counts().plot.pie(startangle=90, ylabel=key, fontsize=15, shadow=True)
    index+= 1
```

# Limpieza de datos

```python
#Queremos evitar tener datos nulos que no le aporten nada de información a nuestro modelo.

#numero de datos por fila y columna antes de la limpieza
print("antes de limpiar, filas y columnas =", data.shape)

#borrando datos nulos tanto en filas como en columnas
dataClean = data.dropna()

#datos despues de la limpieza de datos nulos
print("después de limpiar, filas y columnas =", dataClean.shape)

## No encontramos datos null así que no es necesario implementar una función para rellenar estos valores.
# Si este no fuera el caso, podriamos usar la funcion fillna() la cual rellena esos valores null con el parametro que nosotros le demos a la función.
```

# División de datos

```python
# ahora que limpiamos nuestro datos podemos dividir nuestros datos en dos conjuntos, uno para entrenamiento y el otro para pruebas.

#creo una copia de los datos
dataCopy = data.copy()

#se generan las subclases para las variables categóricas
for key in keyListNonNumeric:
    dms = pd.get_dummies(dataCopy[key], prefix = key)
    #dropeamos las columnas principales
    dataCopy = pd.concat([dataCopy.drop(key, axis = 1), dms], axis = 1)
    
dataCopy.head()
```

```python
#dividimos los datos 
y = dataCopy.target.values
x = dataCopy.drop('target', axis = 1)

trainX, testX, trainY, testY = train_test_split(x, y, test_size = 0.2, stratify = y)
```
